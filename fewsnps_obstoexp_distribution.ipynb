```python
import dr_tools
rsids_conf = dr_tools.loadlist('../RAriskSNPs_20131120_Okada.txt')
rsids_conf2 = rsids_conf[:]
rsids_conf2 += dr_tools.loadlist('../RAriskSNPs_Stahl2010NatGen_Table2and3.txt')
rsids_conf2 += dr_tools.loadlist('../RAriskSNPs_Karlsson2010ARD_Table2.txt')

rsids_conf = list(set(rsids_conf))
rsids_conf2 = list(set(rsids_conf2))
def select(df):
    return df.loc[rsids_conf, :]
def select2(df):
    return df.loc[rsids_conf2, :]
```

```python
from matplotlib import pyplot
import pandas, seaborn, numpy
def load(filename):
    df_new = pandas.read_csv(filename, sep='\t', index_col=0)
    df_new = df_new[df_new['ratio_A0B1']>1.1]
    df_add = pandas.DataFrame({'obs_to_exp':df_new['ratio_A1B1']/df_new['expected_AB_add'], 'expected':['additive' for v in df_new['ratio_A1B1']], 'file':[filename for v in df_new['ratio_A1B1']]})
    df_mult = pandas.DataFrame({'obs_to_exp':df_new['ratio_A1B1']/df_new['expected_AB_mult'], 'expected':['multiplicative' for v in df_new['ratio_A1B1']], 'file':[filename for v in df_new['ratio_A1B1']]})
    #df_mult2 = pandas.DataFrame({'obs_to_exp':df_new['ratio_A1B1']/(df_new['expected_AB_add']+(df_new['OR_A']-1)*(df_new['OR_B']-1)), 'expected':['m2' for v in df_new['ratio_A1B1']], 'file':[filename for v in df_new['ratio_A1B1']]})
    return pandas.concat([df_mult, df_add])
```

```python
df1_2 = pandas.concat([load('sum_actual.txt'), load('dom_actual.txt'), load('rec_actual.txt')])
df2_2 = pandas.concat([load('sum_shuffleAB.txt'), load('dom_shuffleAB.txt'), load('rec_samegroup.txt')])
df3_2 = pandas.concat([load('sum_bgfreqsepgroups.txt'), load('dom_bgfreqsepgroups.txt'), load('rec_bgfreqsepgroups.txt')])
```

```python
def plot(df):
    df['log2_obs_to_exp'] = numpy.log2(df['obs_to_exp']+2**-5)
    df['model'] = df['file'].str.split('_').str[0]
    seaborn.swarmplot(y='obs_to_exp', hue='expected', x='model', data=df, dodge=0.3, alpha=0.4, palette=['orange', 'blue', 'green'], order=['sum', 'dom', 'rec'])
    seaborn.pointplot(y='obs_to_exp', hue='expected', x='model', data=df, ci=95, err_style='bars', legend=False, join=False, markers=['+', '+'], dodge=0.39, color='k', alpha=0.7, order=['sum', 'dom', 'rec'], n_boot=100000, capsize=0.03)
    pyplot.gca().axhline(1)
    pyplot.ylim(0, 3.5)
    #pyplot.gca().get_legend().remove()
```

```python
plot(select(df1_2))
pyplot.show()
plot(select(df2_2))
pyplot.show()
plot(select(df3_2))
pyplot.show()
```
